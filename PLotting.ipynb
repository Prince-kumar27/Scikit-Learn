## 1. Line Plot
> Use: To show trends or relationships between variables over a continuous range.

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y, label="Trend")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

## 2. Scatter Plot
> Use: To visualize relationships or distributions between two sets of data points.

```python
x = [5, 7, 8, 7, 2]
y = [99, 86, 87, 88, 100]
plt.scatter(x, y, color="green")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

## 3. Bar Plot
> Use: To compare categories or discrete data with bars of varying heights.

```python
categories = ['A', 'B', 'C']
values = [4, 7, 3]
plt.bar(categories, values, color='blue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

## 4. Histogram
> Use: To visualize the frequency distribution of a data.


```python
import numpy as np
data = np.random.randn(1000)
#bins=20: Specifies the number of bins (intervals) the data is divided into.
#Larger values create finer granularity; smaller values create coarser grouping.
plt.hist(data, bins=20, color='orange')
plt.title('Histogram')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
x=random.normal(50, 10, 1000)
'''
Mean (loc=50):

The data will have an average value around 50.
Standard Deviation (scale=10):

Most values will fall within ±10 units of the mean (40 to 60).
Sample Size (size=1000):

A total of 1000 random numbers are generated.

'''

plt.hist(x,bins=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.randn(1000)

# Create histogram
plt.hist(data)

# Add title and labels
plt.title('Simple Histogram')
plt.xlabel('Bins')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

## 5. Pie Chart
> Use: To show proportions or percentages of a whole dataset.

```python
sizes = [20, 30, 25, 25]
labels = ['A', 'B', 'C', 'D']
explode = (0.1, 0.1, 0, 0)  # Explode the first slice (A)
#%1.1f%%'->Displays the percentage of each slice on the chart with one decimal place.
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['gold', 'blue', 'red', 'green'],explode=explode,startangle=90)
plt.title('Pie Chart'); plt.show()
```

## 6. Box Plot
> Use: To visualize the spread and outliers in numerical data.

```python
data = [7, 8, 9, 5, 6, 3, 8, 9, 2, 4]
plt.boxplot(data)
plt.title('Box Plot')
plt.ylabel('Values')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load built-in dataset
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
sns.histplot(tips['total_bill'], bins=10, kde=True, color='blue')#kde is the trendline
plt.title('Histogram of Total Bill')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
#hue='sex': This parameter colors the points based on the sex variable.
#Different colors will represent male and female customers.
#style='smoker': This parameter changes the marker style (shape) based on the smoker variable. 
#Points representing smokers and non-smokers will have different marker shapes.
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex')
plt.title('Box Plot of Total Bill by Day')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
plt.title('Violin Plot of Total Bill by Day')
plt.show()
```

```python
sns.pairplot(tips, hue='sex')
plt.show()
```

```python
corr = tips.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=tips, hue='sex')
plt.title('Bar Plot of Total Bill by Day')
plt.show()
```

```python
sns.countplot(x='day', data=tips, hue='sex')
plt.title('Count Plot of Days')
plt.show()
```

```python
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title('Line Plot of Size vs Total Bill')
plt.show()
```

```python
sns.scatterplot(x=tips['total_bill'], y=tips['tip'],hue=tips['sex'])
plt.title('Scatter Plot with Pandas')
# Apply a style
sns.set_style("darkgrid")

# Change color palette
sns.set_palette("pastel")

# Adjust plot size
plt.figure(figsize=(9, 9))

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the Iris dataset
iris = sns.load_dataset("iris")

# Step 2: Calculate the correlation matrix using only numeric columns
corr = iris.select_dtypes(include='number').corr()

# Step 3: Create a heatmap
plt.figure(figsize=(8, 6))  # Optional: Set figure size
sns.heatmap(corr, annot=True, cmap='coolwarm')

# Step 4: Add a title
plt.title('Heatmap of Correlation Matrix (Iris Dataset)')

# Step 5: Show the plot
plt.show()
```

1. Scatter Plot (sns.scatterplot)
Definition: A scatter plot visualizes the relationship between two numerical variables by displaying data points on a two-dimensional plane.
Use Case: To explore potential correlations between variables and identify trends or patterns.

2. Line Plot (sns.lineplot)
Definition: A line plot displays data points connected by lines, often used to visualize trends over time or continuous data.
Use Case: Ideal for showing the relationship between a continuous variable and time or other ordered categories.

3. Bar Plot (sns.barplot)
Definition: A bar plot represents categorical data with rectangular bars, where the length of each bar corresponds to the value it represents.
Use Case: Useful for comparing quantities of different categories or groups.

4. Count Plot (sns.countplot)
Definition: A count plot is a bar plot that shows the count of observations in each categorical bin.
Use Case: Effective for visualizing the frequency of categories in a dataset.

5. Box Plot (sns.boxplot)
Definition: A box plot summarizes the distribution of a dataset by showing its quartiles, median, and potential outliers.
Use Case: Useful for comparing distributions between several groups or identifying outliers.

6. Heatmap (sns.heatmap)
Definition: A heatmap visualizes data in matrix form, using colors to represent values in a 2D space.
Use Case: Ideal for visualizing correlation matrices, comparing the intensity of variables, or displaying complex data.

7. Pair Plot (sns.pairplot)
Definition: A pair plot creates a grid of scatter plots for all pairs of numerical variables in a dataset, along with histograms or density plots along the diagonals.
Use Case: Useful for exploring relationships between multiple variables in a dataset.

8. Histogram (sns.histplot)
Definition: A histogram shows the distribution of a single numerical variable by dividing the data into bins and counting the number of observations in each bin.
Use Case: Useful for understanding the distribution and frequency of data.
